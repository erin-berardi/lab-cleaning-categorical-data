```python
# find all the categorical Data and create a New Data Frame

categorical_df = mark_cust.select_dtypes(include = object)

categorical_df
```

```python
# Check for NaN values and decide what to do with them, do it now.
# There is no Nan Value apparently 

print(categorical_df.isna().sum())
print(mark_cust.isna().sum())
```

```python
# creating a loop for iterating the unique value 
# 
unique_values_list = []  # Create an empty list to store unique values for each column

for column in mark_cust:
    unique_values = mark_cust[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()

```

```python
# creating the same type of loop just for categorical value

unique_values_list = []  

for column in categorical_df:
    unique_values = categorical_df[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()
```

```python
# Checking Dtypes
# and the columns  Effective To Date  could be changed to a numerique value 
# but so far all columns make sens as categorical data except the response one

categorical_df.dtypes
```

```python
# Will clean an wrangle the data as the most appropriate possible 
# change object for gender and response to numerical as there is only two unique value 

categorical_df['Response'] = np.where(categorical_df['Response'] == 'Yes',1,0)

categorical_df['Gender'] = np.where(categorical_df['Gender'] == 'M',1,0)

# could adapt the other data but won t really affect the model 
```

```python
categorical_df['Response']
```

```python
# Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?

# noth columns have the same information almost so we could combine them and keep only the useful info

categorical_df['Policy Type'] = categorical_df['Policy Type'].str.replace('Auto','')

```

```python
categorical_df
```

```python
# dropping the extra Policy column

categorical_df = categorical_df.drop(columns='Policy')
```

```python
# checking now the unique value 

unique_values_list = []  

for column in categorical_df:
    unique_values = categorical_df[column].unique()
    unique_values_list.append((column, unique_values))

# Now unique_values_list contains tuples with column name and unique values
# You can access the unique values for each column as needed
for column_name, unique_values in unique_values_list:
    print(f'Unique values in {column_name}:')
    print(unique_values)
    print()
```

```python
# could separate Education with people over high school and below Higth school
# employed and unemployed ('Medical Leave' 'Disabled' 'Retired')
# location code by ['Suburban' and 'Urban' | 'Rural' ]
# Marital Status married vs single and divorced

# applying all these change in code 

categorical_df['Education'] = categorical_df['Education'].replace({'College': 'High School or Below','Bachelor': 'High School or Below','Master': 'Master or more','Doctor': 'Master or more'})

categorical_df['EmploymentStatus'] = categorical_df['EmploymentStatus'].replace({'Medical Leave': 'Unemployed','Disabled': 'Unemployed','Retired': 'Unemployed'})

categorical_df['Location Code'] = categorical_df['Location Code'].replace({'Suburban': 'Urban'})

categorical_df['Marital Status'] = categorical_df['Marital Status'].replace({'Divorced': 'Single'})

categorical_df
```

```python

# save Data frame as csv file 

categorical_df.to_csv('marketing_cust.csv', index=False)




```
