```python
# Import the necessary libraries
import pandas as pd

import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv('/Users/leozinho.air/Desktop/ironhack_da/class_11/lab-cleaning-categorical-data/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
df
```

```python
# Categorical Data

df_categorical = df.select_dtypes(include = object)

df_categorical
```

```python
# Check for NaN values and decide what to do with them, do it now.

nan_values = df.isnull().sum()
nan_values

df_categorical.duplicated().values.any() # No duplicated
```

```python
# Check all unique values of columns

for i in df_categorical.columns:
    print(f'The unique values in {i} are: {df_categorical[i].unique()}')
```

```python
# Check dtypes. Do they all make sense as categorical data?

for i in df_categorical.columns:
    print(f'The datatype of {i} is: {df_categorical[i].dtype}') # Obviously all columns are obj
    
# There are some columns that no makes sense as categorical. -> e.g. Response/Gender can be converted to binary
                                                                #    Effective To Date can be converted to date-time
                                                                    
```

```python
# Does any column contain alpha and numeric data?  Decide how to clean it and do it now.

# Loop to display the columns that have mixed alpha/num data
for column in df_categorical.columns:
    true_alphanum = False
    for index,value in df_categorical[column].items():
        if any(_.isalpha() for _ in str(value)) & any(_.isdigit() for _ in str(value)):
            true_alphanum = True
            break
    if true_alphanum:
        print(f'Column {column} contains mixed alphanumeric data.')
```

```python
# 1. Customer

df_categorical['Customer'].nunique() # 9134 

# Each value in the 'Customer' column is unique and likely serves as an identifier
# There's no need to clean or modify this column since each entry appears to represent a different customer.
```

```python
# 2. Policy

df_categorical['Policy'].nunique() # 9
df_categorical['Policy'].unique() #'Corporate L3', 'Personal L3', 'Corporate L2', 'Personal L1',
                                  #'Special L2', 'Corporate L1', 'Personal L2', 'Special L1','Special L3'
       
# If the values in this column accurately represent distinct types or levels of insurance policies and are properly categorized,
# no further cleaning or modifications for this column may be necessary due to its limited and well-defined unique values.
```

```python
# 3. Renew Offer Type

df_categorical['Renew Offer Type'].nunique() # 4
df_categorical['Renew Offer Type'].unique() # 'Offer1', 'Offer3', 'Offer2', 'Offer4'

# This suggests that this column likely represents different types or categories of renewal offers provided to customers. 
# no further cleaning is necessary
```

```python
# Would you choose to do anything else to clean or wrangle the categorical data? 

# 1. Converting Response to binary - Gender values to dummies

df_categorical['Response'] = np.where(df_categorical['Response'] == 'Yes', 1,0).astype(int)

# 2. One-hot encoding over Gender

df_categorical = pd.get_dummies(df_categorical, columns=['Gender'])

# 3. Converting Effective to Date to datetime type

df_categorical['Effective To Date'] = pd.to_datetime(df_categorical['Effective To Date'], format='%m/%d/%y')

df_categorical
```

```python
# Compare policy_type and policy.
# What information is contained in these columns
# Can you identify what is important?

columns = ['Policy Type', 'Policy']
for i in df_categorical[columns]:
    print(f'The unique values in {i} are: {df_categorical[i].unique()}')

# Policy Type seems to categorize the broad types or categories of insurance policies.
# Policy gives more detailed information about the specific levels or tiers within each policy type.

```

```python
# Check number of unique values in each column, can they be combined in any way to ease encoding? 

# Policy Type could be a higher-level category and Policy a more detailed breakdown within each type.
# Therefore i can potentially combine the two column in a new one.
# Then i will encode the two original columns with One-Hot encoding and Label econding

# 1. Combine the Policy and Policy type columns
df_categorical['Policy Combined'] = df_categorical['Policy Type'] + '_' + df_categorical['Policy'].apply(lambda x: x.split(' ')[1])
df_categorical.drop(columns=['Policy'], inplace=True)

# 2. One-Hot encoding over Policy type

policy_type_encoded = pd.get_dummies(df_categorical['Policy Type'])

# Concatenate the one-hot encoded Policy Type with the original dataframe
df_categorical = pd.concat([df_categorical, policy_type_encoded], axis=1)

df_categorical.drop('Policy Type', axis=1, inplace=True)

# Standardize columns name

cols = []
for i in range(len(df_categorical.columns)):
    cols.append(df_categorical.columns[i].lower().replace(' ', '_')) 
df_categorical.columns = cols

```

```python
# Saving the cleaned df in a .csv file 

# Define the filename for the CSV file
filename = 'categorical.csv'

# Save the DataFrame to a CSV file
df_categorical.to_csv(filename, index=True)
```
