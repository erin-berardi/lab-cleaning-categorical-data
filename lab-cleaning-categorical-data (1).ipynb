```python
#Function to extract only numeric values and aplying the function in 'renew_offer_type' column.

def keep_numeric(series):
    return series.str.replace(r'\D', '', regex=True)

data_cat['renew_offer_type'] = keep_numeric(data_cat['renew_offer_type'])
data_cat['renew_offer_type'] = pd.to_numeric(data_cat['renew_offer_type'], errors='coerce')
data_cat
```
