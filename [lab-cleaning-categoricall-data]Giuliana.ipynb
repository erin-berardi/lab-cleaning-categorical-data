```python
# Does any column contain alpha and numeric data? Decide how to clean it and do it now.
# There are no column with alpha and numeric data

def alpha_and_numeric(df):
    for column in df.columns:
        for value in df[column]:
            if value.isalpha() and value.isnumeric():
                return column  

    return 'There are none'  

# Example usage with categorical_df
result = alpha_and_numeric(categorical_df)
print("Contains alpha and numeric data:", result)

```
