```python
#1. Get the data and clean table:
df = pd.read_csv(r'C:\Users\ljant\Desktop\Ironhack\06_Advanced_Statistics\lab-cleaning-categorical-data\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv')

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df)
# 9134 x 24
```

```python
#1. Find and save the categorical columns

data_types = df.dtypes
print (data_types)

def column_types (data):
    num_col = df.select_dtypes(include=['int', 'float']).columns
    cat_col = df.select_dtypes(include=['object']).columns
    return num_col, cat_col

df_num = df.select_dtypes('number')
df_cat = df.select_dtypes('object')
# 9134 x 16
```

```python
#2. Deal with NaNs
sum_na = df_cat.isna().sum()
print (sum_na)
#-> no NaNs in the df
```

```python
#3. Check for unique values

def unique_values(dataset):
    values = {}
    for column in dataset.columns:
        values[column] = dataset[column].unique()
    return values

values = unique_values(df_cat)

for column, value in values.items():
    print(f'Unique values in column "{column}":')
    for v in value:
        print(v)
```

```python
#4. Check the datatype
data_types = df_cat.dtypes
print (data_types)

'''
The column 'effective_to_date' should be a datetime type and not an object
'''

df_cat['effective_to_date'] = pd.to_datetime(df_cat['effective_to_date'])
data_types = df_cat.dtypes
print (data_types)
```

```python
#5. Transforming suitable columns into numeric columns and store them in a new df (dropping the rest)
df_cat_adj = df_cat.drop(['customer', 'effective_to_date', 'policy_type'] , axis = 1)

df_cat_adj = pd.get_dummies(df_cat_adj, dtype = int)

def header_function (data):
    cols = []
    for i in range (len(data.columns)):
        cols.append(data.columns[i].lower().replace(' ','_'))
    
    data.columns = cols

    return data

header_function(df_cat_adj)
# 9134 x 54

'''
I chose to transform every possibile important categorical column into a numeric one (with the exception of 'customer',
'effective_to_date','policy_type', because here are no useful information stored.)

This new df can now be concated with the numeric df creating a new df ready for regression models (if the numeric data is
cleaned as well.)
'''
```
