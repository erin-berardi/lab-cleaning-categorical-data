```python
# 1. Import the necessary libraries load the data and start a new notebook. 
# Using the same data as the previous lab: we_fn_use_c_marketing_customer_value_analysis.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Find all of the categorical data. Save it in a categorical_df variable.
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')

new_columns = []
for col in customer_df.columns:
    new_columns.append(col.lower().replace(' ', '_'))
    
customer_df.columns = new_columns
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
categorical_df = customer_df.select_dtypes(object)
print(categorical_df.shape)
categorical_df.head()
```

```python
# 3. Check for NaN values and decide what to do with them, do it now.

categorical_df.isna().sum()*100/len(categorical_df)

# There are no NaN values in this dataset
```

```python
# 4. Check all unique values of columns.

for col in categorical_df.columns:
    unique_values = categorical_df[col].unique()
    print(f"Unique values in {col}: {unique_values}")
```

```python
# 5. Check dtypes. Do they all make sense as categorical data?
categorical_df.dtypes

# Yes, all columns make sense as categorical data. 
```

```python
# 6. Does any column contain alpha and numeric data? Decide how to clean it and do it now.

# Yes, the 'customer', 'policy' and 'renew_offer_type' columns contain alphanumeric values.
# The 'customer' column will contain a large number of unique values and it will not provide much information that can be used for encoding, therefore it could be dropped. 
# The 'policy' column contains the same information as the 'policy_type' column and therefore it could also be dropped.

cat_df = categorical_df.copy()
cat_df = cat_df.drop(['policy','customer'], axis=1)
cat_df.head()
```

```python
# 7. Would you choose to do anything else to clean or wrangle the categorical data? Comment your decisions and do it now.
# No, so far I will keep all the columns with categorical data.
```

```python
# 8. Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?
# See answer in question number 6. The 'policy' column was dropped as it contains the same information as 'policy_type'.
```

```python
# 9. Check number of unique values in each column, can they be combined in any way to ease encoding? 
# Comment your thoughts and make those changes.
for col in cat_df.columns:
    unique_values = cat_df[col].unique()
    print(f"Unique values in {col}: {unique_values}")
```

```python
# 10. Save the cleaned catagorical dataframe as categorical.csv You will use this file again this week.

cat_df.to_csv('categorical.csv', index=False)
```
