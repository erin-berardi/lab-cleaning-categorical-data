# 6. Does any column contain alpha and numeric data? Decide how to clean it and do it now.

Yes, the customer column contains an alpha and numeric data

```python
categorical_df_clean = categorical_df.copy()
```

```python
categorical_df_clean['customer_id'] = categorical_df_clean['Customer'].str.extract('(\d+)')
```

```python
categorical_df_clean['Customer'] = categorical_df_clean['Customer'].str.replace('\d+', '')
```
