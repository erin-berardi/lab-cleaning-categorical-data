```python
#importing csv
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

mkt = pd.read_csv(r"C:\Users\pedro\Desktop\Ironhack\lab-cleaning-categorical-data\files_for_lab\we_fn_use_c_marketing_customer_value_analysis.csv")
mkt.head(3)
```

```python
#Finding all of the categorical data and saving it in a dataframe

mkt_categoricals=mkt[[i for i in mkt.columns if mkt.dtypes[i]=='object']]
#mkt_categoricals = mkt.select_dtypes(['object'])

mkt_categoricals.head(3)
```

```python
#function checking for NA

def check_NA(df):
    print(df.isna().sum().sum())
    print("\n",df.isna().sum() / len(df))
```

```python
#applying the function
#There are no NA's in the dataframe

check_NA(mkt)
```

```python
#Checking how many unique values are in each column

mkt_categoricals.describe()[1:2]
```

```python
#Checking dtypes
#Variable 'customer' is an ID and 'Effective To Date' is a date

mkt_categoricals.dtypes
```

```python
#Transforming 'Effective To Date' to datetime

mkt_categoricals['Effective To Date']=pd.to_datetime(mkt_categoricals['Effective To Date'])
mkt_categoricals.dtypes
```

```python
mkt_categoricals = mkt_categoricals.drop('Effective To Date',axis=1)
mkt_categoricals.head(3)
```

```python
#function standardizing headers names

def standardize_headers(df):
    #standardizing headers names
    cols = []
    for i in range(len(df.columns)):
        cols.append(df.columns[i].lower().replace(' ','_'))
    df.columns = cols   
    return df
```

```python
standardize_headers(mkt_categoricals)
```

```python
#One hot enconding

mkt_categoricals_encoded = pd.get_dummies(mkt_categoricals.drop(['customer'],axis=1))
mkt_categoricals_encoded = pd.concat([mkt_categoricals['customer'], mkt_categoricals_encoded], axis=1)
mkt_categoricals_encoded
```
